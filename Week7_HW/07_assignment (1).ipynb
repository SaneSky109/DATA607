## 3. Fix missing and invalid values in data.

#### Identify missing rows

```python
sunshine.isnull().any(axis = 1).sum()
```

```python
nan_rows  = sunshine[sunshine.isnull().any(axis=1)]

nan_rows
```

#### Drop rows with too much missing data

Row index 25 has many missing data points that are too difficult or impossible to identify. For example, I do not know the city and have no way to replace it. I am also missing the total so I can't really replace the missing months using the Yearly column. This row can be dropped.

```python
sunshine.drop(25, axis = 0, inplace = True)

nan_rows  = sunshine[sunshine.isnull().any(axis=1)]

print(sunshine.isnull().any(axis = 1).sum())
nan_rows
```

#### Use deductive reasoning to replace NaN values 

The remaining missing values can be replaced using deductive reasoning:
* Countries can be identified via the city
* Single missing month values can be found by subtracting all other months from Yearly
* Double missing month values can be found  by subtracting all other months from Yearly and split the difference between the two
* Yearly can be found by adding all other months together

As shown below all NaN values have been accounted for.

```python
### Country column replacements ###

# Algiers is the capital of Algeria
sunshine.loc[3,"Country"] = 'Algeria'

# Canberra is the capital of Australia
sunshine.loc[16,"Country"] = 'Australia'


### Month column replacements ###

# Get month totals
month = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec']
month_sum = nan_rows[month].sum(numeric_only = True, axis = 1)

# Subtract month total from Yearly
diff = nan_rows['Yearly'] - month_sum

# find number of months with missing values
num_nan = nan_rows[month].isnull().sum(axis=1)

# Now we have enough information to replace NaN values

sunshine.loc[3,"Jan"] = diff[3] / num_nan[3]
sunshine.loc[3,"Feb"] = diff[3] / num_nan[3]
sunshine.loc[16,"Yearly"] = month_sum[16]
sunshine.loc[39,"Oct"] = diff[39] / num_nan[39]
sunshine.loc[46,"Mar"] = diff[46] / num_nan[46]
sunshine.loc[53,"Sep"] = diff[53] / num_nan[53]

nan_rows  = sunshine[sunshine.isnull().any(axis=1)]

print(sunshine.isnull().any(axis = 1).sum())
nan_rows
```

## 4. Create new columns based on existing columns or calculations.

I was curious to see the amount of sun each city recieved on a daily basis, on average. To create these variables, I took the value for each month and divided it by the appropriate number of days. See function below.

```python
pd.options.mode.chained_assignment = None  # default='warn'

sunshine_non_numeric = sunshine[['Country','City']] 

def get_daily_sunshine(df):
    
    # Get column names
    columns = list(sunshine)
    
    for i in columns:
        if 'Jan' in  i:
            df['Jan_daily_sunshine'] = df['Jan'] / 31
        elif 'Feb' in  i:
            df['Feb_daily_sunshine'] = df['Feb'] / 28
        elif 'Mar' in  i:
            df['Mar_daily_sunshine'] = df['Mar'] / 31
        elif 'Apr' in  i:
            df['Apr_daily_sunshine'] = df['Apr'] / 30
        elif 'May' in  i:
            df['May_daily_sunshine'] = df['May'] / 31
        elif 'Jun' in  i:
            df['Jun_daily_sunshine'] = df['Jun'] / 30
        elif 'Jul' in  i:
            df['Jul_daily_sunshine'] = df['Jul'] / 31
        elif 'Aug' in  i:
            df['Aug_daily_sunshine'] = df['Aug'] / 31
        elif 'Sep' in  i:
            df['Sep_daily_sunshine'] = df['Sep'] / 30
        elif 'Oct' in  i:
            df['Oct_daily_sunshine'] = df['Oct'] / 31
        elif 'Nov' in  i:
            df['Nov_daily_sunshine'] = df['Nov'] / 30
        elif 'Dec' in  i:
            df['Dec_daily_sunshine'] = df['Dec'] / 31
        elif 'Yearly' in  i:
            df['Yearly_daily_sunshine'] = df['Yearly'] / 365
        
    return df

# function works on numeric data only
numeric_data = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec', 'Yearly']
sunshine1 = get_daily_sunshine(sunshine[numeric_data])

# column bind non numeric and numeric data together 
sunshine = pd.concat([sunshine_non_numeric, sunshine1], axis=1, ignore_index=False)

sunshine.head()
```

## 11. Group your dataset by one column, and get the mean, min, and max values by group.

### Groupby()

The groupby function will combine all cities in a particular country into a singular groups (Country). 

```python
sunshine.groupby(by='Country').describe()
```

### .agg() and .apply()

The agg function aggregates the data using one or more operations. Below shows the data grouped by country and aggrated by mean.

```python
sunshine.groupby(by='Country')['Yearly_daily_sunshine'].agg('mean')
```

The .apply function is used to apply functions to columns in the dataframe. Below I created a column to assign if a city recieves a large amount of sunlight. Most cities recieved below 8 hours of sunlight as a yearly daily average.

```python
def assign_level(column):
    if column < 8:
        return 'low sunlight'
    elif column >= 8 and column <= 16:
        return 'avg. sunlight'
    elif column > 16:
        return 'high sunlight'
    
sunshine['sunshine_level'] = sunshine['Yearly_daily_sunshine'].apply(assign_level)

sunshine['sunshine_level']
```

```python
sunshine.groupby(by = 'sunshine_level')['sunshine_level'].agg('count')
```

```python
sunshine.groupby(by = 'Country')['Country'].agg('count').sort_values(ascending=False)
```

## 12. Group your dataset by two columns and then sort the aggregated results within the groups.

Grouping by two columns requires that the sorted values function take a tuple as the olumn becomes multilevel.

```python
q12 = sunshine.groupby(['Country', 'sunshine_level']).agg({'Jan': ['mean', 'min', 'max'],
                                                    'Feb': ['mean', 'min', 'max'],
                                                    'Mar': ['mean', 'min', 'max'],
                                                    'Apr': ['mean', 'min', 'max'],
                                                    'May': ['mean', 'min', 'max'],
                                                    'Jun': ['mean', 'min', 'max'],
                                                    'Jul': ['mean', 'min', 'max'],
                                                    'Aug': ['mean', 'min', 'max'],
                                                    'Sep': ['mean', 'min', 'max'],
                                                    'Oct': ['mean', 'min', 'max'],
                                                    'Nov': ['mean', 'min', 'max'],
                                                    'Dec': ['mean', 'min', 'max']})
# need to provide tuple as it is a multi level column
q12.sort_values([('Jan', 'mean')], ascending=[True])
```
